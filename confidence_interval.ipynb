```python
import pandas as pd
import numpy as np
import scipy.stats
#https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data
```

```python
df = pd.read_excel('/Volumes/Untitled/Desktop/grant_work/twitter-code/bayes/bug_testing_v2/by_hashtag/solar_counties_v2.xls', sheetname='solar_counties_v2')
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
mean_confidence_interval(df['compound'])
```

```python
ci = df['compound'].groupby(df['NAME']).apply(mean_confidence_interval)
```

```python
df = pd.DataFrame(ci)
```

```python
filename = '/Volumes/Untitled/Desktop/grant_work/twitter-code/bayes/bug_testing_v2/by_hashtag/solar_counties_confidence_interval.csv'
 
df.to_csv(filename, index=True, encoding='utf-8')
```
