```python
#https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe
path = '/Volumes/Untitled/Desktop/grant_work/working/twitter-code/bayes/data/bug_testing/20180424_20180519_geocoded'
all_files = glob(os.path.join(path, "*.json"))
df_from_each_file = (pd.read_json(f, lines=False) for f in all_files)
df = pd.concat(df_from_each_file)
```

```python
df = df.reset_index(drop=True)
```

```python
filename = '/Volumes/Untitled/Desktop/grant_work/working/twitter-code/bayes/data/bug_testing/20180424_20180519_geocoded.csv'

df.to_csv(filename, index=False, encoding='utf-8')
```
