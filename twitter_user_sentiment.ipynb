```python
import os
import time
import datetime

from collections import Counter
import numpy as np
import pandas as pd
from nltk.corpus import stopwords

from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *

from nltk.sentiment.vader import SentimentIntensityAnalyzer
import unicodedata

import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
```

```python
directory = '/Users/davidleifer/Desktop/grant_work/working/twitter-code/user-tweets/user-json-data'
#directory = os.fsencode(data)

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".csv") or filename.endswith(".jsonl"): 
        #print(os.path.join(directory, filename))
        print("Begin processing " + filename + " at",datetime.datetime.now().strftime("%H:%M:%S"))
        df = pd.read_json(os.path.join(directory, filename), lines=True)
        
        
        print("Cleaning " + filename + " at",datetime.datetime.now().strftime("%H:%M:%S"))

        #remove major stopwords
        stop = stopwords.words('english')
        df['tweet_without_stopwords'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
        #df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace(r'(?:\@|https?\://)\S+', '')

        #remove periods
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[.]','')
        #remove commas
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[,]','')
        #remove -
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[-]','')
        #remove @
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[@]','')
        #remove &
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[&]','')
        #remove variety of characters
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[1]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[2]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[3]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[4]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[5]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[6]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[7]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[8]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[9]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[10]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[25]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[26]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace(']','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[I]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[Te]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[S]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[$]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[A]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[0]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[*]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[::]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[-]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[7th]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[(]','')
        df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace('[)]','')

        
        #remove hashtag_list
        #df['tweet_without_stopwords'] = df['tweet_without_stopwords'].str.replace(hashtag_list[i],'')

        #count frequency
        count = Counter(" ".join(df["tweet_without_stopwords"]).split()).most_common(50)
        count = pd.DataFrame(count)
        count.columns = ['Word', 'Count']
        
        #create bar chart of frequency 
        X = count['Word']
        Y = count['Count'] 
        y_pos = np.arange(len(X))
        fig_size = plt.rcParams["figure.figsize"]
        fig_size[0] = 12
        fig_size[1] = 12
        plt.rcParams["figure.figsize"] = fig_size
        plt.barh(y_pos, Y, align='center', alpha=0.5)
        plt.yticks(y_pos, X)
        plt.xlabel('Word Count')
        plt.title('Word Frequency for ' + filename)
        plt.savefig('/Users/davidleifer/Desktop/grant_work/working/twitter-code/user-tweets/twitter-user-bar-histo/' + filename + '-bar.png')
        plt.show()
        
        tweets = df['tweet_without_stopwords']
        df = pd.DataFrame(tweets)
        df["compound"] = ''
        df["neg"] = ''
        df["neu"] = ''
        df["pos"] = ''
        sid = SentimentIntensityAnalyzer()
        print("Start sentiment analysis " + filename + " at",datetime.datetime.now().strftime("%H:%M:%S"))

        for user, row in df.T.iteritems():
            try:
                sentence = unicodedata.normalize('NFKD', df.loc[user, 'tweet_without_stopwords'])
                ss = sid.polarity_scores(sentence)
                df.set_value(user, 'compound', ss['compound'])
                df.set_value(user, 'neg', ss['neg'])
                df.set_value(user, 'neu', ss['neu'])
                df.set_value(user, 'pos', ss['pos'])
            except TypeError:
                print(df.loc[user, 'tweet_without_stopwords'])
    
        #save to file
        sent_analysis = '/Users/davidleifer/Desktop/grant_work/working/twitter-code/user-tweets/user-sentiment-analysis/' + filename + '.csv'
        df.to_csv(sent_analysis, index=False, encoding='utf-8')
    
        fig_size = plt.rcParams["figure.figsize"]
        fig_size[0] = 12
        fig_size[1] = 9
        plt.rcParams["figure.figsize"] = fig_size
        x = df['compound']
        plt.hist(x, normed=True, bins=30)
        plt.ylabel('Frequency');
        plt.title('Sentiment Frequency for ' + filename)
        plt.savefig('/Users/davidleifer/Desktop/grant_work/working/twitter-code/user-tweets/twitter-user-bar-histo/' + filename + '-histo.png')
        plt.show()

        continue
    else:
        continue
        
print("Finished Processing" ,datetime.datetime.now().strftime("%H:%M:%S"))
```
